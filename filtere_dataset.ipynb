```python
filtered_pgn = open('/home/abhishek/Desktop/result.pgn')
random_moves = []
for game in filtered_pgn:
    game = chess.pgn.read_game(filtered_pgn)
    board = game.board()
    
    num_moves = len(list(game.mainline_moves())) -10
# white wins the game
    if game.headers['Result'] == '1-0':
        if(num_moves <= 10):
            continue
        
        r_num = random.choice(range(10, num_moves, 2))
    
        count = 0
        for move in game.mainline_moves():
            
            if(count != r_num):
                count+=1
                
            else:
                #print(move)
                print(r_num)
                #print(game.headers)
                print(board.san(move))
                print(board.fen())
                random_moves.append((board.san(move),board.fen(),str(move)))
                break
            board.push(move)
# black wins            
    if game.headers['Result'] == '0-1':
        if(num_moves <= 11):
            continue
            
        r_num = random.choice(range(11, num_moves, 2))
        count = 0
        for move in game.mainline_moves():
            if(count != r_num):
                count+=1   
            else:
                #print(move)
                print(r_num)
                #print(game.headers)
                print(board.san(move))
                print(board.fen())
                random_moves.append((board.san(move),board.fen(),str(move)))
                break
            board.push(move)
```

```python
print(random_moves)
header = ['San','Fen','UCI']
with open('dataset.csv','a') as fcsv:
    writer = csv.writer(fcsv)
    writer.writerow(header)
    writer.writerows(random_moves)
```
